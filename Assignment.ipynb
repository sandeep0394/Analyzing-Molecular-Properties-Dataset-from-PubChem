```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

url = "https://pubchem.ncbi.nlm.nih.gov/rest/pug/compound/cid/2244,1983,702,6322,5957/property/MolecularWeight,XLogP,InChIKey,CanonicalSMILES/CSV"

df = pd.read_csv(url)
print(df.head())
```

```python
2.Inspect the data: Check column names, data types, and handle any missing values.
```

```python
# Display basic information about the dataset
print("Column Names:", df.columns.tolist())

# Check data types of each column
print("\nData Types:\n", df.dtypes)

# Check for missing values in each column
print("\nMissing Values:\n", df.isnull().sum())

# Show basic statistics
print("\nDescriptive Statistics:\n", df.describe())

# Display the first few rows to visually inspect
df.head()
```

```python
3. Perform basic data cleaning if needed.
```

```python
 # Remove duplicates if any
df = df.drop_duplicates()

# Confirm the shape after removing duplicates
print(f"Data shape after removing duplicates: {df.shape}")

```

```python
# Plot distributions of key columns to check for outliers
plt.figure(figsize=(10, 5))

# Distribution of Molecular Weight
plt.subplot(1, 2, 1)
sns.histplot(df['MolecularWeight'], kde=True, color='red')
plt.title('Distribution of Molecular Weight')

# Distribution of XLogP
plt.subplot(1, 2, 2)
sns.histplot(df['XLogP'], kde=True, color='yellow')
plt.title('Distribution of XLogP')

plt.tight_layout()
plt.show()
```

```python

# Check for empty or malformed entries in 'CanonicalSMILES' and 'InChIKey'
print("CanonicalSMILES column null values:", df['CanonicalSMILES'].isnull().sum())
print("InChIKey column null values:", df['InChIKey'].isnull().sum())

# Check for invalid entries in XLogP (like NaN, infinite, etc.)
print("Invalid XLogP values:", df['XLogP'].isnull().sum() + (df['XLogP'] == float('inf')).sum() + (df['XLogP'] == float('-inf')).sum())
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

# Plot Molecular Weight vs XLogP (scatter plot)
plt.figure(figsize=(8, 6))
sns.scatterplot(x='MolecularWeight', y='XLogP', data=df, color='blue')
plt.title('Molecular Weight vs XLogP')
plt.xlabel('Molecular Weight (g/mol)')
plt.ylabel('XLogP (LogP value)')
plt.grid(True)
plt.tight_layout()
plt.show()

# Plot XLogP vs CID (bar plot)
plt.figure(figsize=(8, 6))
sns.barplot(x='CID', y='XLogP', data=df, hue='CID', palette='viridis', legend=False)
plt.title('XLogP vs CID')
plt.xlabel('CID (Compound ID)')
plt.ylabel('XLogP (LogP value)')
plt.xticks(rotation=45)
plt.grid(True)
plt.tight_layout()
plt.show()
```

```python
6. Explain how trends in these molecular properties can support solvent selection in drug development.
```

```python
Understanding molecular properties like molecular weight and LogP is essential when selecting suitable solvents during drug development. Compounds with lower molecular weight 
and lower LogP values tend to be more water-soluble, making them better suited for aqueous solvents. In contrast, molecules with higher molecular weight or higher LogP are 
typically more lipophilic, requiring organic or non-polar solvents for effective dissolution. For example, a hydrophilic drug with a low LogP may dissolve well in water-based 
systems, whereas a more hydrophobic compound with a higher LogP would be more compatible with organic solvents. These characteristics directly influence a drugs solubility, 
absorption rate, and overall bioavailability, all of which are critical for designing effective formulations.
```
